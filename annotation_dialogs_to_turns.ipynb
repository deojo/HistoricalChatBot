```python
"""
Created on Thu Jun 2 09:54:19 2023
@author: DSJoshi
"""

!pip install datasets

import datasets
from datasets import load_dataset
import pandas as pd
from datasets import Dataset, DatasetDict

def generate_dialog_turns(df):
    outputs = []
    for i, record in df.iterrows():
            annotation = record['annotation_text']
            lines = annotation.split('\n')
            for line in lines:
                line = line.strip()
                if len(line) > 0:
                    speaker_idx = line.find(":")
                    speaker = line[0:speaker_idx].strip()
                    if speaker not in ['Student', 'Teacher']: continue
                    annotation_start_idx = line.rfind('[')
                    annotation_end_idx = line.rfind(']')
                    line_annotation= ''
                    text = ''
                    if annotation_start_idx >= 0 and annotation_start_idx <= len(line) and annotation_end_idx >= 0 and annotation_end_idx <= len(line):
                        line_annotation = line[annotation_start_idx+1:annotation_end_idx].strip().replace(',','|')
                        text = line[speaker_idx+1:annotation_start_idx].strip()
                        outputs.append({'id':record['id'], 'speaker':speaker, 'text': text, 'annotation': line_annotation})
    return pd.DataFrame(outputs)

ds = load_dataset('Deojoandco/reddit-ah-dialog-annotations')
train_df = pd.DataFrame(ds['train'])
val_df = pd.DataFrame(ds['validation'])
test_df = pd.DataFrame(ds['test'])

dialogturns_train_df = generate_dialog_turns(train_df)

dialogturns_val_df = generate_dialog_turns(val_df)
dialogturns_test_df = generate_dialog_turns(test_df)

dialogturns_ds = datasets.DatasetDict({
    'train': Dataset.from_pandas(dialogturns_train_df),
    'validation': Dataset.from_pandas(dialogturns_val_df),
    'test': Dataset.from_pandas(dialogturns_test_df)
})

print(dialogturns_ds)

dialogturns_ds.push_to_hub('reddit-ah-dialogturns-annotations', token ='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')
```
